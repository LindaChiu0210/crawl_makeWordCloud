```python
def get_ptt_feedback(website_path):
    s = requests.Session()
    r = s.get(website_path, cookies={'over18': '1'})
    #將原始碼做整理
    soup = BeautifulSoup(r.text, 'lxml')
    #使用find_all()找尋特定目標
    articles = soup.find_all('div', 'push')
    #印出
    #print(articles)
    for article in articles:
        #去除掉冒號和左右的空白
        messages = article.find('span','f3 push-content').getText().replace(':','').strip()
        text_container.append(messages)
        #f.write(messages + "\n")
```

```python
def read_text_output_wordcloud(text_box,output_wordcloud_path):
    temp_text=text_box
    total_num=len(temp_text)
    print('共有',total_num,'篇文章。')
    
    jieba.set_dictionary(r'C:\Users\linda\Downloads\0410\dict.txt.big.txt')
    jieba.load_userdict(r'C:\Users\linda\Downloads\0411\user_0410_utf8.txt')
    with open(r'C:\Users\linda\Downloads\0411\stopWord.txt', 'r', encoding='utf8') as f:  
        stops = f.read().split('\n') 
    temp_terms_container=[]
    
    print('開始處理字串...')
    for i in range(total_num):
        temp_terms = [t for t in jieba.cut(temp_text[i], cut_all=False) if t not in stops]
        temp_terms_container.extend(temp_terms)
    
    terms=[]
    for i in range(len(temp_terms_container)):
        if (len(temp_terms_container[i]))>1: 
            terms.append(temp_terms_container[i])
            
    print('字串處理完畢。')
    font = r'C:\Users\linda\Downloads\0410\msjh.ttc'
    my_wordcloud = WordCloud(background_color="white",
                         #mask=mask,
                         #colormap='Set2',
                         font_path=font,
                         width=2400, height=2400, 
                         min_font_size=20,
                         margin=2,
                         prefer_horizontal=1
                        )
    
    print('開始產生文字雲...')
    my_wordcloud.generate_from_frequencies(frequencies=Counter(terms))
    plt.figure( figsize=(20,10), facecolor='k')
    plt.imshow(my_wordcloud,interpolation='bilinear')
    plt.axis("off")
    plt.tight_layout(pad=0)
    
    print('將文字雲顯示並存檔。')
    plt.savefig(output_wordcloud_path)
    plt.show()
```
